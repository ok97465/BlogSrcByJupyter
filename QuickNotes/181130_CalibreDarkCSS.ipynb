# Calibre

## Linux에서 한글 쓰기

Calibre는 자체 입력기를 사용하기 때문에 한글 입력을 위해서는 Linux 입력기를 Calibre 폴더에 복사해야 한다.

```sh
cp /usr/lib/x86_64-linux-gnu/qt5/plugins/platforminputcontexts/libqt5im-nimf.so /opt/calibre/plugins/platforminputcontexts
```
